```python
def get_token():
    #Authentication
    scope = 'user-library-read'
    token = util.prompt_for_user_token(username, scope, client_id=cid, client_secret=secret, redirect_uri=redirect_uri)

    if token:
        sp = spotipy.Spotify(auth=token)
        return token
    else:
        print("Can't get token for", username)
```

```python
#all features and keys should be lowercase with underscore instead of spaces
def get_audiofeatures(url,query_limit,cycle,show=True):
    features = ['danceability','energy','key','loudness','mode','speechiness','acousticness','instrumentalness',
                'liveness','valence','tempo','time_signature']
    #get url
    try:
        resp = requests.get(url=url,
                            headers={'Authorization': 'Bearer '+ get_token()})
        resp.raise_for_status()
        
    except requests.exceptions.HTTPError as err:
        print(err)
        
        
    #response in json format
    response = resp.json()
    
    for x in range(query_limit):
        try:
            #dictionary 
            audiofeatures = {}
            track_uri = response['audio_features'][x]['uri'].split(':')
            audiofeatures.update({'track_uri':track_uri[2]})
            master_track_audiofeatures.append(audiofeatures)
        
            for track_feature in features:
                feature_value = response['audio_features'][x][track_feature]
                audiofeatures.update({track_feature:feature_value}) #ties track_uri and features
            
        except:
            print(cycle,x,url)
    
    return True
    
    #if show:
    #    print('URI: {uri} - AudioFeatures: {af}'.format(uri=track_uri,af=audiofeatures))
```

```python
i = 0
master_track_audiofeatures=[]
start = time.time()

#Convert df into list
df = pd.read_csv('Data/artist_album_track_uri.csv',sep='|')
list_track_uri = df['track_uri'].tolist()

#Spotify_track_uri_limit - amount of uris allowed to be in endpoint
#Cycle - is how many times around the loop
spotify_track_uri_limit = 100
cycle = math.ceil((len(list_track_uri)/spotify_track_uri_limit))

#Split list_track_uri into chunks by cycle amount
track_uri_chunks = numpy.array_split(numpy.array(list_track_uri),cycle)
track_uri_chunks

for i in range(cycle):
        print('Cycle: ',i)
        char = '%2C'
        string_track_uri_chunks = char.join(track_uri_chunks[i])
        track_audiofeatures = get_audiofeatures('https://api.spotify.com/v1/audio-features?ids={id}'
                                                .format(id=string_track_uri_chunks),spotify_track_uri_limit,cycle)
        
stop = time.time()
duration = (stop - start)/60
```

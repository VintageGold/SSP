```python
def getTracks(nexturl):
    track_dict = {
        'URI'
    }
    success = 'Success'
    try:
        resp = requests.get(url=nexturl,
                            headers={'Authorization': 'Bearer ' + token})
        resp.raise_for_status()
    except requests.exceptions.HTTPError as err:
        print(err)
        
    response = resp.json()
    
    #responsetotal
    #response 
    
    for x in range(50):
        try:
            URI = response['items'][x]['track']['uri']
            Track = response['items'][x]['track']['name']
            
            print(x,URI,'-',Track)
        except:
            return nexturl 
        
    if (nexturl!= None):
        getTracks(response['next'])
    else:
        return nexturl
```

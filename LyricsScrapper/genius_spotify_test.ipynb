```python
import lyricsgenius
import string
import time
```

```python
def _removeNonAscii(s):
    return "".join(i for i in s if ord(i)<128)
```

```python
headers = ["artist","song","lyrics"]
spotify_list = []
rapgenius_list = []
music_dbfile = "Data/lowercase_Final.txt"
fn = "lowercase_song.txt"
lyrics = 0
counter = 0
delimiter = '|'

def get_lyrics(song=0,artist=0):
    genius = lyricsgenius.Genius("CLMJDhT2HYCbiXT9grKoh3LZ2hjiCRuvYYxTOj3lARJgM03frK2k739jf9UwDxFs",
remove_section_headers=True
    )
    #Make all lines one string
    lyrics =str(genius.search_song(song,artist,get_full_info=False).lyrics.splitlines())
    #Remove punctuation
    lyrics = _removeNonAscii(lyrics.translate(str.maketrans('','', string.punctuation)))
    return lyrics
start = time.time()
with open(music_dbfile,'r') as musicfile:
    next(musicfile) #Skip header row
    head = [next(musicfile) for x in range(2000)]

    with open(fn, "w+") as outfile:
        for line in head:
            current = line.split('-')
            try:
                lyrics = get_lyrics(artist=current[0],song=current[2])
                outfile.write(current[0]+delimiter)
                outfile.write(current[2]+delimiter)
                outfile.write(lyrics)
                outfile.write('\n')
            except:
                continue
                counter += 1
    end = time.time()
    duration = str((end - start))
```
